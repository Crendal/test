```python
from sklearn.model_selection import train_test_split  # X : feature값
X=iris.data
y=iris.target
X_train,               X_test,            y_train,         y_test = train_test_split( X, y, test_size=0.33, random_state=42)
# test size 학습 train 비율 7:3으로 나누겠다
#학습셋의 피처    평가셋의 피처     학습셋의 레이블    평가셋의 레이블
```
